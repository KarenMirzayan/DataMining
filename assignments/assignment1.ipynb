Objective: Build a preprocessing pipeline to automate the data preparation process.

```python
X_train
```

```python
df1 = pd.read_csv("Titanic-Dataset.csv")
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1 = df1.drop(columns=["Name", 'PassengerId', 'Ticket', 'Cabin', 'Embarked'])
df1
```

```python
df1["Family_Size"] = df1["SibSp"] + df1['Parch'] + 1
df1 = df1.drop(columns=["SibSp", "Parch"])
df1
```

```python
df1.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Normalizer

X = df1.drop('Survived', axis=1)
y = df1['Survived']   
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

categorical_cols = ['Sex', 'Pclass'] 
numerical_cols = ['Age', 'Fare', 'Family_Size']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='mean')),
            ('norm', Normalizer())
        ]), numerical_cols),
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ]), categorical_cols)
    ]
)

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

pipeline.fit(X_train, y_train)
B = pipeline.transform(X_test)

print(*B)
```

```python
from sklearn.linear_model import LogisticRegression

pipeline =  Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

What are the benefits of using a preprocessing pipeline? A: Lowers amount of work

How does the pipeline ensure consistency between training and test data transformations? A: in preprocessing stage we use simple math so output should be consistent. In model fit-transforming stage accuracy may vary

How can you extend the pipeline to include additional preprocessing steps? A: Add more steps to pipeline, we can add any class that has fit-transform.
